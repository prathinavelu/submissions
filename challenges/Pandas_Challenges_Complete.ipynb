```python
import pickle

with open('movie_data_dict_yearly.pkl','r') as picklefile:
    movie_data = pickle.load(picklefile)
    
while None in movie_data:
    movie_data.remove(None)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import collections

%matplotlib inline
```

```python
movies_clean = []
for movie in movie_data:
    if movie['rating'] == 'GP':
        movie['rating'] = 'PG'
    if movie['domestic_total_gross'] <> 'N/A' and movie['release date'] <> 'N/A' and movie['runtime'] <> 'N/A':
        movies_clean.append(movie)

movies_all_data = pd.DataFrame(movies_clean)

print len(movies_all_data)
```

```python
#Challenge 2: plot runtime vs domestic total gross

gross_by_date = sorted(zip(movies_all_data['runtime'],movies_all_data['domestic_total_gross']))

#some outliers removed for plot
x = [runtime for runtime, gross in gross_by_date[:-15]]
y = [gross for runtime, gross in gross_by_date[:-15]]


plt.figure(figsize=(15, 5))
plt.plot(x,y,'*')

plt.title('Domestic Gross vs Runtime')
plt.xlabel('Runtime (Minutes)')
plt.ylabel('Box Office Domestic Gross')
```

```python
#Challenge 3: group your data by Rating and find the average runtime and domestic total gross at each level of Rating
movies_by_rating = movies_all_data.groupby('rating')

pd.options.display.float_format = '{:,.0f}'.format
movies_by_rating.mean()
```

```python
#Challenge 4: make one figure with (N=the number of MPAA ratings there are) subplots, 
#and in each plot the release date vs the domestic total gross

#remove unrated and not yet rated for plots
movies_by_some_ratings = movies_all_data.drop(movies_by_rating.get_group('Unrated').index)
movies_by_some_ratings = movies_by_some_ratings.drop(movies_by_rating.get_group('Not Yet Rated').index)
movies_by_some_ratings_grouped = movies_by_some_ratings.groupby('rating')

for i, group in movies_by_some_ratings_grouped:
    plt.figure(figsize=(10,2.5))
    group.plot(x='release date', y='domestic_total_gross', title=str(i))
```

```python
#Challenge 5: what director in your dataset has the highest gross per movie?
movies_by_director = movies_all_data.groupby('director')
movies_by_director.mean().sort('domestic_total_gross', ascending=False)
```

```python
#Challenge 6: bin your dataset into years (if applicable) and make a bar graph with error bars of gross each year

movies_by_year = movies_all_data.groupby(movies_all_data['release date'].map(lambda x : x.year))


errors_gross = movies_by_year['domestic_total_gross'].sem()
errors_runtime = movies_by_year['runtime'].sem()
movies_by_year['domestic_total_gross'].mean().plot(figsize= (20,7.5),yerr =errors_gross,kind = 'bar',title = 'Avg Domestic Total Gross by Year')

```
