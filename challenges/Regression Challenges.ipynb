```python
%matplotlib inline
import pandas as pd
import numpy as np
from datetime import datetime
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt
from statsmodels.sandbox.regression.predstd import wls_prediction_std
from sklearn.linear_model import LinearRegression

# Note: Here I'm loading my movie csv file to make a data frame
# You can work on your own data frame

movies = pd.read_csv('2013_movies.csv', parse_dates={'Date':['ReleaseDate']})
```

```python
import math

movies_with_budget = movies[ -np.isnan(movies.Budget) ]
movies_with_budget['Ones'] = 1.0
movies_with_budget['Log_Gross'] = movies_with_budget['DomesticTotalGross'].apply(lambda x: math.log(x))
#movies_with_budget['Log_Budget'] = movies_with_budget['Budget'].apply(lambda x: math.log(x))
movies_with_budget.head()
```

```python
X = movies_with_budget[['Budget', 'Ones']]
Y = movies_with_budget['Log_Gross']
linmodel = sm.OLS(Y,X).fit()
linmodel.summary()
```

```python
#Challenge 1

#Instead of predicting total gross directly, build a linear model that predicts the log (logarithm) of the total gross. 
#Plot the model. Plot the predictions versus actual values. Does this look like a better model?

plt.scatter(movies_with_budget.Budget, movies_with_budget.Log_Gross)
plt.xlabel('Budget')
plt.ylabel('Box Office')
plt.plot(movies_with_budget.Budget, linmodel.predict(X), 'r')

prstd, conf_lower, conf_upper = wls_prediction_std(linmodel)
plt.plot(movies_with_budget['Budget'], conf_lower, 'y--')
plt.plot(movies_with_budget['Budget'], conf_upper, 'y--')
```

```python
#Challenge 2

#We are fitting and checking the predictions on the exact same dataset! Divide your data into two sets: 
#a training and a test set (roughly 75% training, %25 test). Fit a model on the training set, check the predictions 
#(by plotting versus actual values) in the test set.


movies_with_budget_sample = movies_with_budget.reindex(np.random.permutation(movies_with_budget.index))
movies_with_budget_training = movies_with_budget_sample[:int(len(movies_with_budget_sample)*.75)]
movies_with_budget_test = movies_with_budget_sample[int(len(movies_with_budget_sample)*.75):]


```

```python
X = movies_with_budget_training[['Budget', 'Ones']]
Y = movies_with_budget_training['Log_Gross']
linmodel_new = sm.OLS(Y,X).fit()
linmodel_new.summary()
```

```python
plt.scatter(movies_with_budget_test.Budget, movies_with_budget_test.Log_Gross)

plt.xlabel('Budget')
plt.ylabel('Box Office')
plt.plot(movies_with_budget_training.Budget, linmodel_new.predict(X), 'r')
```

```python
X = movies_with_budget_test[['Budget', 'Ones']]
Y = movies_with_budget_test['Log_Gross']
linmodel_new = sm.OLS(Y,X).fit()

plt.plot(movies_with_budget_test['Log_Gross'], linmodel_new.predict(X), 'go')

plt.plot(movies_with_budget_test['Log_Gross'], movies_with_budget_test['Log_Gross'], 'b--')

plt.xlabel('Actual Gross')
plt.ylabel('Predicted Gross by Rating Model')
```

```python
#Challenge 3

#Build a model that uses average director gross as a feature (one of the predictor variables). Fit and evaluate. 
#Would you use this model to predict budgets of upcoming movies? We will discuss

movies_with_budget_and_director = movies_with_budget[pd.notnull(movies_with_budget['Director'])]
grouped = movies_with_budget.groupby('Director').mean()


movies_with_budget_and_director['DirectorAvgGross'] = [grouped['DomesticTotalGross'][director] for 
                                                       director in movies_with_budget_and_director.Director]

movies_with_budget_and_director['Log_DirectorAvgGross'] = [math.log(grouped['DomesticTotalGross'][director]) for 
                                                       director in movies_with_budget_and_director.Director]

#movies_with_budget.head()
```

```python
X = movies_with_budget_and_director[['Budget', 'Log_DirectorAvgGross', 'Ones']]
Y = movies_with_budget_and_director['Log_Gross']
linmodel_d = sm.OLS(Y,X).fit()
linmodel_d.summary()
```

```python
plt.scatter(movies_with_budget_and_director.Budget, movies_with_budget_and_director.Log_Gross)
plt.xlabel('Budget')
plt.ylabel('Box Office')
plt.plot(movies_with_budget_and_director.Log_Budget, linmodel_d.predict(X), 'r')

prstd, conf_lower, conf_upper = wls_prediction_std(linmodel_d)
plt.plot(movies_with_budget_and_director['Budget'], conf_lower, 'y--')
plt.plot(movies_with_budget_and_director['Budget'], conf_upper, 'y--')
```
